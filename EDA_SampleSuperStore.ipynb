Deleting the variable postal code

```python
col = ['Postal Code']
df = data.drop(columns = col, axis = 1)
df.head()
```

#Visualization

```python
plt.figure(figsize = (16,8))
sns.countplot(y=data['Ship Mode'], data = df)
plt.show()
```

People ordered the most from the Standard Class Shipping mode

```python
plt.figure(figsize = (16,8))
sns.countplot(y=data['Segment'], data = df)
plt.show()
```

Most people are from the Consumer segment

```python
plt.figure(figsize = (16,8))
sns.countplot(y = data['Sub-Category'], data= df)
```

Binders and papers are the most in-demand items. Whereas copies are least in demand

```python
plt.figure(figsize = (16,8))
sns.countplot(y=data['Region'], data = df)
```

Majorly the people are from the West region

```python
plt.figure(figsize = (16,8))
sns.countplot(y=data['Category'], data = df)
```

Most sales are from the office supplies

```python
plt.figure(figsize = (16,8))
plt.bar('Category', 'Region', data = df)
```

It is equally distributed

```python
plt.figure(figsize = (16,8))
plt.bar("Sub-Category", 'Category', data = df)
plt.xlabel('Sub-Category')
plt.ylabel('Category')
plt.show()
```

Most of the products are from office supplies category

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(16,8))
sns.countplot(df['State'], data = df)
plt.xticks(rotation=90)
plt.show()
```

Major purchases are from California. Followed by NY and Texas
